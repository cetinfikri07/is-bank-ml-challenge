```python
from pathlib import Path
from warnings import simplefilter

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#from matplotlib import rcParams
import seaborn as sns

simplefilter('ignore')
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)

plt.style.use('ggplot')
plt.rc('figure',autolayout=True, figsize=(11,4))
plt.rc(
    "axes",
    labelweight="bold",
    labelsize="large",
    titleweight="bold",
    titlesize=18,
    titlepad=10
)
```

```python
data_dir = Path('../data')

train = pd.read_parquet(
    data_dir / 'train_final.parquet'
)

test = pd.read_parquet(
    data_dir / 'test_final.parquet'
)

ss = pd.read_parquet(
    data_dir / 'submission_sample_final.parquet'
)
```

```python
train.head()
```

#### Normalize N Seconds

```python
train['n_seconds'] = train[['n_seconds_1','n_seconds_2','n_seconds_3']].sum(axis=1)


for col in ['n_seconds_1','n_seconds_2','n_seconds_3']:
    train[col] = train[col] / train['n_seconds']


train.head()
```

#### Binarize Target

```python
def binarize_target(row, menu_cnt=9):
    menu_label = np.zeros(menu_cnt).astype(int)
    for i, menu in enumerate(row["target"].split(", ")):
        
            menu_id = int(menu.replace("menu", "")) -1
            menu_label[menu_id] = 1
        
    return menu_label

train["binarize_target"] = train.apply(binarize_target, axis=1)

train.head()
```

```python
def regression_target(row, menu_cnt=9):
    menu_label = np.zeros(menu_cnt)
    for i, menu in enumerate(row["target"].split(", ")):
        menu_id = int(menu.replace("menu", "")) -1
        menu_label[menu_id] = row[f"n_seconds_{i+1}"]
        
    return menu_label

train['regression_target'] = train.apply(regression_target,axis=1)

train.head()
```

```python
def change_carrier(carrier):
    if carrier not in ["TURKCELL", "VODAFONE TR", "TURK TELEKOM"]:
        carrier = "OTHER"
        
    return carrier


def change_devicebrand(devicebrand):
    if devicebrand != "Apple":
        devicebrand = "Android"
        
    return devicebrand

train['carrier'] = train['carrier'].apply(change_carrier)
train['devicebrand'] = train['devicebrand'].apply(change_devicebrand)

train.head()
```

```python
train['strata'] = train['devicebrand'] + train['carrier']

train.head()
```

```python
from sklearn.model_selection import train_test_split

train_df, test_df = train_test_split(
    train,
    test_size=0.2,
    random_state=0,
    stratify=train['strata']
)

feature_cols = [col for col in train_df.columns if "feature_" in col]
feature_cols.remove("feature_48")
feature_cols.remove("feature_49")
```

```python
import numpy as np
from sklearn.multioutput import MultiOutputRegressor, MultiOutputClassifier
from sklearn.linear_model import Ridge, LogisticRegression, LinearRegression
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import jaccard_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import QuantileTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

regression = False
if regression:
    target_col = 'regression_target'
    model = MultiOutputRegressor(LinearRegression())
else:
    target_col = 'binarize_target'
    model = MultiOutputClassifier(LogisticRegression())

qt = QuantileTransformer(n_quantiles=10, random_state=0)
ohe = OneHotEncoder(handle_unknown='ignore')
count_cols = ['feature_48','feature_49']
ohe_cols = ['carrier','devicebrand']
seconds_cols = ['n_seconds_1','n_seconds_2','n_seconds_3']

cols = feature_cols + ohe_cols + seconds_cols + count_cols
ohe_cols_index = [index for index, col in enumerate(cols) if col in ohe_cols]
count_cols_index = [index for index, col in enumerate(cols) if col in count_cols]

# features and labels 
X_train = train_df[cols]
y_train = np.concatenate(train_df[target_col].values).reshape(-1,9)

# ct pipeline

ct = ColumnTransformer([
    ('quantile', qt, count_cols_index),
    ('ohe', ohe, ohe_cols_index),
], remainder='passthrough', n_jobs=-1)

pipe = Pipeline([
    ('ct',ct),
    ('model', model)
])

fitted_pipe = pipe.fit(X_train,y_train)
```

```python
def binarize_pred(preds, menu_cnt=9):
    menu_label = ["0"] * menu_cnt
    indices = np.argsort(preds)[::-1][:3]
    
    for index in indices:
        menu_label[index] = "1"
        
    return "".join(menu_label)


temp_df = train_df.copy()

temp_df['preds'] =  np.hstack([i[:,1].reshape(-1,1) for i in fitted_pipe.predict_proba(temp_df[cols])]).tolist()

temp_df['preds'] = temp_df['preds'].apply(binarize_pred)
temp_df['list_preds'] = temp_df['preds'].apply(list)
temp_df["list_preds"] = temp_df["list_preds"].apply(lambda x: [int(i) for i in x])

y_true = np.array(temp_df["binarize_target"].tolist())
y_pred = np.array(temp_df["list_preds"].tolist())
js = jaccard_score(y_true, y_pred, average='micro')

temp_df.head()
```
